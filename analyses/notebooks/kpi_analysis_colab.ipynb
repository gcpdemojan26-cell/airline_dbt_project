# KPI Analysis (Colab)
This notebook demonstrates reading dbt models from BigQuery and computing/visualizing KPIs.

```python
# Install and import
!pip install --quiet google-cloud-bigquery pandas matplotlib
from google.cloud import bigquery
import pandas as pd
import matplotlib.pyplot as plt
project_id = 'YOUR_GCP_PROJECT_ID'
dataset = 'staging'
client = bigquery.Client(project=project_id)
```

```python
# Query monthly revenue generated by dbt
query = f"SELECT month, revenue FROM `{project_id}.{dataset}.monthly_revenue` ORDER BY month"
df = client.query(query).to_dataframe()
df.head()
```

```python
# Plot
plt.figure(figsize=(10,4))
plt.plot(df['month'], df['revenue'], marker='o')
plt.title('Monthly Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.grid(True)
plt.show()
```
